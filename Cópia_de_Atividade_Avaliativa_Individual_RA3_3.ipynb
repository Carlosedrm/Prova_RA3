# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
# O valor de x com os y's
x = np.linspace(0, 2, 1000)
y1 = x * np.exp(-x**2)
y2 = x + 1

# As linhas
plt.plot(x, y1, label='$y=xe^{-x^{2}}$', color='blue')
plt.plot(x, y2, label='$y = x + 1$', color='red')

# As labels
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)  # Habilitar as linhas de grade

# Preenchendo a área entre as curvas
plt.fill_between(x, y1, y2, where=y1 < y2, interpolate=True, color='green', alpha=0.3)

# Mostrando o gráfico
plt.legend()
plt.show()
```

```python
x1 = sy.Symbol('x')
y1 = x1*sy.exp(-x1**2)
y2 = x1 + 1

area = sy.integrate(y1 - y2, (x1, 1, 2))

print(f"A área entre as duas funções é {area:.3f}")
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



**A**

```python
import numpy as np
import matplotlib.pyplot as plt

# Função original
def original_function(x):
    return x**2 + 1

# Função integrada
def integrated_function(x):
    return (1/3)*x**3 + x

# Limites de integração
a = 0
b = 2

# Gerar valores de x e y para o gráfico
x = np.linspace(-1, 3, 100)  # Intervalo um pouco maior para visualização
y_original = original_function(x)
y_integrated = integrated_function(x)

# Plotar o gráfico da função original
plt.plot(x, y_original, label='Função Original')

# Plotar o gráfico da função integrada
plt.plot(x, y_integrated, label='Função Integrada')

# Destacar os limites de integração
plt.fill_between(x, y_integrated, where=[(xi >= a and xi <= b) for xi in x], alpha=0.3)

# Adicionar linhas verticais nos limites de integração
plt.axvline(a, color='r', linestyle='--', label='Limite Inferior')
plt.axvline(b, color='g', linestyle='--', label='Limite Superior')

# Configurações adicionais do gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico das Funções Original e Integrada')
plt.legend()

# Exibir o gráfico
plt.grid()
plt.show()
```
